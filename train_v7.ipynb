```python
import keras
import pandas as pd
from keras.layers.core import Dense, Dropout
from keras.layers.recurrent import GRU
from keras.models import Sequential, load_model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import copy

from keras.layers.recurrent import GRU
from keras.models import Model
from keras.layers import Input, Convolution2D, Flatten, Dense, Concatenate,Dropout
import copy
import numpy as np
import tensorflow as tf

true_stand_data=np.load(r'.//datatarget//true_array_target.npy')
false_stand_data=np.load(r'.//datatarget//false_array_target.npy')

true_stand_data=true_stand_data[:,:,1:12]
false_stand_data=false_stand_data[:,:,1:12]

sample_num={'true_num':true_stand_data.shape[0],'false_num':false_stand_data.shape[0]}
class_weight = {
                0: (1 / sample_num['false_num'] * (sample_num['true_num'] + sample_num['false_num'])) / 2,
                1: (1 / sample_num['true_num'] * (sample_num['true_num'] + sample_num['false_num'])) / 2
                }

true_sample_channel_1 = []
true_sample_channel_2 = []
true_sample_channel_3 = []
true_sample_channel_4 = []
true_sample_channel_5 = []
true_sample_channel_6 = []
true_sample_channel_7 = []
true_sample_channel_8 = []
true_sample_channel_9 = []
true_sample_channel_10 = []
true_sample_channel_11 = []

false_sample_channel_1 = []
false_sample_channel_2 = []
false_sample_channel_3 = []
false_sample_channel_4 = []
false_sample_channel_5 = []
false_sample_channel_6 = []
false_sample_channel_7 = []
false_sample_channel_8 = []
false_sample_channel_9 = []
false_sample_channel_10 = []
false_sample_channel_11 = []

for sample_number in range(true_stand_data.shape[0]):
    true_sample_channel_1.append(copy.deepcopy(true_stand_data[sample_number,:,0]))
    true_sample_channel_2.append(copy.deepcopy(true_stand_data[sample_number,:,1]))
    true_sample_channel_3.append(copy.deepcopy(true_stand_data[sample_number,:,2]))
    true_sample_channel_4.append(copy.deepcopy(true_stand_data[sample_number,:,3]))
    true_sample_channel_5.append(copy.deepcopy(true_stand_data[sample_number,:,4]))
    true_sample_channel_6.append(copy.deepcopy(true_stand_data[sample_number,:,5]))
    true_sample_channel_7.append(copy.deepcopy(true_stand_data[sample_number,:,6]))
    true_sample_channel_8.append(copy.deepcopy(true_stand_data[sample_number,:,7]))
    true_sample_channel_9.append(copy.deepcopy(true_stand_data[sample_number,:,8]))
    true_sample_channel_10.append(copy.deepcopy(true_stand_data[sample_number,:,9]))
    true_sample_channel_11.append(copy.deepcopy(true_stand_data[sample_number,:,10]))

for sample_number in range(false_stand_data.shape[0]):
    false_sample_channel_1.append(copy.deepcopy(false_stand_data[sample_number,:,0]))
    false_sample_channel_2.append(copy.deepcopy(false_stand_data[sample_number,:,1]))
    false_sample_channel_3.append(copy.deepcopy(false_stand_data[sample_number,:,2]))
    false_sample_channel_4.append(copy.deepcopy(false_stand_data[sample_number,:,3]))
    false_sample_channel_5.append(copy.deepcopy(false_stand_data[sample_number,:,4]))
    false_sample_channel_6.append(copy.deepcopy(false_stand_data[sample_number,:,5]))
    false_sample_channel_7.append(copy.deepcopy(false_stand_data[sample_number,:,6]))
    false_sample_channel_8.append(copy.deepcopy(false_stand_data[sample_number,:,7]))
    false_sample_channel_9.append(copy.deepcopy(false_stand_data[sample_number,:,8]))
    false_sample_channel_10.append(copy.deepcopy(false_stand_data[sample_number,:,9]))
    false_sample_channel_11.append(copy.deepcopy(false_stand_data[sample_number,:,10]))
    
combin_train_data1 = np.concatenate((true_sample_channel_1,false_sample_channel_1),axis=0)
combin_train_data2 = np.concatenate((true_sample_channel_2,false_sample_channel_2),axis=0)
combin_train_data3 = np.concatenate((true_sample_channel_3,false_sample_channel_3),axis=0)
combin_train_data4 = np.concatenate((true_sample_channel_4,false_sample_channel_4),axis=0)
combin_train_data5 = np.concatenate((true_sample_channel_5,false_sample_channel_5),axis=0)
combin_train_data6 = np.concatenate((true_sample_channel_6,false_sample_channel_6),axis=0)
combin_train_data7 = np.concatenate((true_sample_channel_7,false_sample_channel_7),axis=0)
combin_train_data8 = np.concatenate((true_sample_channel_8,false_sample_channel_8),axis=0)
combin_train_data9 = np.concatenate((true_sample_channel_9,false_sample_channel_9),axis=0)
combin_train_data10 = np.concatenate((true_sample_channel_10,false_sample_channel_10),axis=0)
combin_train_data11 = np.concatenate((true_sample_channel_11,false_sample_channel_11),axis=0)

combin_target_data= np.concatenate((np.ones((sample_num['true_num'],1)),np.zeros((sample_num['false_num'],1))),axis=0)
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
combin_target_data=ohe.fit_transform(combin_target_data).toarray()

# Model 1
in_1 = Input(shape=(110,1))
x_100_1 = GRU(units=100)(in_1)
x_10_1 =Dense(10, activation='tanh')(x_100_1)

# Model 2
in_2 = Input(shape=(110,1))
x_100_2 = GRU(units=100)(in_2)
x_10_2 =Dense(10, activation='tanh')(x_100_2)

# Model 3
in_3 = Input(shape=(110,1))
x_100_3 = GRU(units=100)(in_3)
x_10_3 =Dense(10, activation='tanh')(x_100_3)

# Model 4
in_4 = Input(shape=(110,1))
x_100_4 = GRU(units=100)(in_4)
x_10_4 =Dense(10, activation='tanh')(x_100_4)

# Model 5
in_5 = Input(shape=(110,1))
x_100_5 = GRU(units=100)(in_5)
x_10_5 =Dense(10, activation='tanh')(x_100_5)

# Model 6
in_6 = Input(shape=(110,1))
x_100_6 = GRU(units=100)(in_6)
x_10_6 =Dense(10, activation='tanh')(x_100_6)

# Model 7
in_7 = Input(shape=(110,1))
x_100_7 = GRU(units=100)(in_7)
x_10_7 =Dense(10, activation='tanh')(x_100_7)

# Model 8
in_8 = Input(shape=(110,1))
x_100_8 = GRU(units=100)(in_8)
x_10_8 =Dense(10, activation='tanh')(x_100_8)

# Model 9
in_9 = Input(shape=(110,1))
x_100_9 = GRU(units=100)(in_9)
x_10_9 =Dense(10, activation='tanh')(x_100_9)

# Model 10
in_10 = Input(shape=(110,1))
x_100_10 = GRU(units=100)(in_10)
x_10_10 =Dense(10, activation='tanh')(x_100_10)

# Model 10
in_11 = Input(shape=(110,1))
x_100_11 = GRU(units=100)(in_11)
x_10_11 =Dense(10, activation='tanh')(x_100_11)


total_out_concat = Concatenate(axis=-1)([x_10_1,x_10_2,x_10_3,x_10_4,x_10_5,x_10_6,x_10_7,x_10_8,x_10_9,x_10_10,x_10_11])
model_final_dense_out = Dense(2, activation='softmax')(total_out_concat)

model = Model(inputs=[in_1,in_2,in_3,in_4,in_5,in_6,in_7,in_8,in_9,in_10,in_11], outputs=model_final_dense_out)
print(model.summary())

import os
if 'GRU_predict_v7.h5' in os.listdir('./'):
    model = load_model("./GRU_predict_v7.h5")

callbacks_list = [    
    keras.callbacks.EarlyStopping(
        monitor='accuracy', 
        patience=30
    ),
    # 保存模型
    keras.callbacks.ModelCheckpoint(
    filepath = 'GRU_predict_v7.h5', 
    monitor='loss', 
    save_best_only=True),
    
    keras.callbacks.ReduceLROnPlateau(
        monitor='loss', 
        factor=0.8,
        patience=5, 
        mode='auto',
        min_lr=0.00003)
]


model.compile(loss='categorical_crossentropy', #continu together
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              metrics=['accuracy']
             
             )

model.fit(x=[combin_train_data1,combin_train_data2,combin_train_data3,combin_train_data4,combin_train_data5,combin_train_data6,combin_train_data7,combin_train_data8,combin_train_data9,combin_train_data10,combin_train_data11],
          y=combin_target_data,
          batch_size=360,
          epochs=100,
          shuffle=True,
          verbose=1,
          class_weight=class_weight,
          callbacks=callbacks_list      
         )

```
